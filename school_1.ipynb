### Sürekli Zaman (CT) & Ayrık Zaman (DT) Sinyali

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Sürekli Zaman (Continuous Time - CT), zamanın kesintisiz olarak devam ettiği bir sistemde, zamanın herhangi bir noktasında sinyal değerinin belirlenebilmesi anlamına gelir. Sürekli zaman sinyalleri, zaman ekseninde herhangi bir aralıkta tanımlı olan fonksiyonlardır. Bu sinyaller matematiksel olarak genellikle analitik ifadelerle temsil edilir, yani fonksiyonlar bir denklemle tanımlanır ve her zaman diliminde bir değer alabilir.
Örneğin;
$$
x(t) = 3 \cdot \sin(0.1 \cdot \pi \cdot t) + 2 \cdot \cos(0.05 \cdot \pi \cdot t)
$$

```python
t = np.linspace(-30, 30, 1000)
# 3 * sin(0.1 * pi * t) + 2 * cos(0.05 * pi * t)
x_t = 3 * np.sin(0.1 * np.pi * t) + 2 * np.cos(0.05 * np.pi * t) 
plt.stem(t, x_t, label='Continues Time (CT)')
plt.xlabel(xlabel='t')
plt.ylabel(ylabel='x(t)')
plt.show()
```

Ayrık zamanlı sinyallerde, zaman aralığı belirli noktalarla sınırlıdır. 
Bu noktalar genellikle eşit aralıklarla seçilir.
Fonksiyonlar sadece bu belirli zaman noktalarında tanımlıdır.
Formülde bu belirli noktalar için hesaplamalar yapılır ve her bir noktada sinyalin değeri belirlenir. 
Son olarak, bu noktalar grafik üzerinde çizilir.
$$
x[n] = 2 \cdot \sin\left(\frac{\pi}{4} n\right)
$$

```python
n = np.arange(-3, 4, 1)
x_n = 2 * np.sin(np.pi / 4 * n)

plt.stem(n, x_n, label="x(n)")
plt.title(label='Discrete Time')

plt.xlabel(xlabel='n')
plt.ylabel(ylabel='x(n)')
plt.legend()
plt.show()
```

### ÇİFT FONKSİYON

Çift fonksiyonlar (Even Functions), matematiksel bir özellik gösteren fonksiyonlardır. Bu özellik, fonksiyonun zamanla simetrik olması anlamına gelir. Yani, bir fonksiyon **f(t)**, aşağıdaki koşulu sağlıyorsa **çift fonksiyon** olarak kabul edilir:

$$
f(t) = f(-t) \quad \text{her} \quad t \in \mathbb{R}
$$

Bu özellik, fonksiyonun **yatay simetri** göstermesi anlamına gelir. Çift fonksiyonlar genellikle sinyal işleme ve sistem analizlerinde önemlidir, çünkü bu tür sinyallerin Fourier dönüşümleri gibi bazı matematiksel işlemlerle özel ilişkileri vardır.

### Örnek:
Örnek olarak, **t^2** fonksiyonu çift fonksiyonudur, çünkü:

$$
f(t) = t^2 \quad \text{ve} \quad f(-t) = (-t)^2 = t^2
$$

Başka bir örnek de **cos(t)** fonksiyonudur, çünkü:

$$
\cos(t) = \cos(-t)
$$

Ancak **sin(t)** fonksiyonu **tek fonksiyon**dur, çünkü:

$$
\sin(t) = -\sin(-t)
$$

### Çift Fonksiyonların Zamanla İlişkisi

Çift fonksiyonlar zamanla simetrik olduğundan, örnekleme noktaları üzerinden yapılan çizimler veya işlemler bu simetrik yapıyı yansıtır. Sürekli veya ayrık zaman sinyalleriyle bu fonksiyonlar, genellikle sistem yanıtlarını anlamada veya sinyal işleme alanında kullanılır.

#### Çift Fonksiyonun Ayrık Zamanlı Gösterimi:
Ayrık zaman sinyali için de aynı simetri uygulanabilir. Örneğin, bir ayrık zaman sinyali **x[n] = n^2** çift fonksiyonudur çünkü:

$$
x[n] = x[-n]
$$

Bu özellik, ayrık zaman sinyalleri için sistem yanıtlarının daha kolay hesaplanmasına olanak sağlar.

### Uygulama:
Aşağıdaki Python kodu, bir çift fonksiyon olan **f(t) = t^2**'yi sürekli zamanlı bir sinyal olarak çizmektedir:

```python
import numpy as np
import matplotlib.pyplot as plt

# Zaman dizisi
t = np.linspace(-10, 10, 1000)

# Çift fonksiyon
f_t = t**2

plt.plot(t, f_t, label='f(t) = t^2', linewidth=2)
plt.title(label='Çift Fonksiyon')
plt.grid(True)
plt.xlabel(xlabel='t')
plt.ylabel(ylabel='f(t)')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend(loc='upper center')
plt.show()
```

### Tek Fonksiyon (Odd Function) Nedir?

**Tek Fonksiyonlar**, aşağıdaki koşulu sağlayan fonksiyonlardır:

$$
f(t) = -f(-t) \quad \text{her} \quad t \in \mathbb{R}
$$

Bu koşul, fonksiyonun **yatayda simetrik olmadığını** ancak **çift taraflı ters simetrik** olduğunu belirtir. Başka bir deyişle, tek fonksiyonlar zamanla tersine evrilir, yani fonksiyon değerleri negatif zaman için tam tersi olur.

### Örnekler:

1. **sin(t)** fonksiyonu tek fonksiyondur, çünkü:

$$
\sin(t) = -\sin(-t)
$$

2. **t^3** fonksiyonu da tek fonksiyondur, çünkü:

$$
f(t) = t^3 \quad \text{ve} \quad f(-t) = (-t)^3 = -t^3
$$

### Tek Fonksiyonların Zamanla İlişkisi

Tek fonksiyonlar, zamanla tersine simetrik bir yapı gösterir. Yani, bir zaman noktasındaki fonksiyon değeri, zamanın negatifindeki fonksiyon değeriyle tam zıt olur. Bu özellik, sinyal işleme ve sistem analizi gibi alanlarda özellikle önemli olabilir çünkü bir sistemin tepkisi, zamanla ters bir şekilde evrilebilir.

#### Tek Fonksiyonun Ayrık Zamanlı Gösterimi:

Ayrık zamanlı sinyallerde de aynı simetri geçerlidir. Örneğin, bir ayrık zaman sinyali **x[n] = n^3** tek fonksiyondur çünkü:

$$
x[n] = -x[-n]
$$

### Python Uygulaması:

Aşağıdaki Python kodu, bir tek fonksiyon olan **f(t) = sin(t)**'yi sürekli zamanlı bir sinyal olarak çizmektedir:

```python
import numpy as np
import matplotlib.pyplot as plt

# Zaman dizisi
t = np.linspace(-10, 10, 1000)

# Tek fonksiyon: f(t) = x^3
f_t = t**3

plt.figure(figsize=(10, 6))
plt.plot(t, f_t, label="f(t) = t^3", color='g', linewidth=2)
plt.title('Tek Fonksiyon: f(t) = t^3')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid(True)
plt.show()
```
